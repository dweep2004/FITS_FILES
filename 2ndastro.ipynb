```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
M32=fits.open("frame-u-003367-3-0075.fits")

print(M32)
```

```python
data=M32[0].data

plt.imshow(data, cmap="gray")
plt.colorbar()
plt.show()
```

```python
histogram=plt.hist(data.flat, bins=500)
```

```python
from matplotlib.colors import LogNorm

plt.imshow(data, cmap="gray", norm=LogNorm())
plt.colorbar()
plt.show()
```

```python
M32u=fits.open('frame-u-003367-3-0075.fits')
M32i=fits.open('frame-i-003367-3-0075.fits')
M32g=fits.open('frame-g-003367-3-0075.fits')
```

```python
from astropy.visualization import make_lupton_rgb
```

```python
g=M32g[0].data
u=M32u[0].data
i=M32i[0].data
```

```python
rgb_default= make_lupton_rgb(i, g, u, stretch=2.5, Q=5)


plt.imshow(rgb_default, origin="lower")
plt.savefig("rgbM42.jpg")
```
